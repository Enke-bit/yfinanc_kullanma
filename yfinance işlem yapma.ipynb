## Verileri görselleştirme.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Veriyi resetleyerek 'Date' kolonunu kullanmak
data_apple.reset_index(inplace=True)

# Kapanış fiyatlarını çizim
plt.figure(figsize=(14, 7))
sns.lineplot(x='Date', y='Close', data=data_apple)
plt.title('AAPL Stock Price Over Time')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.xticks(rotation=45)
plt.show()
```

```python
# Seaborn stil ayarlarını kullanma.
sns.set(style="darkgrid")

plt.figure(figsize=(14, 7))
sns.lineplot(x='Date', y='Close', data=data_apple, marker='o')  # Fiyatın denk geldiği yere nokta atmak.
plt.title('AAPL Stock Price Over Time')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.xticks(rotation=45)
plt.show()
```

```python
# Hacim ve Kapanış Fiyatını Birlikte Görselleştirme

fig, ax1 = plt.subplots(figsize=(14, 7))

# Fiyat grafiği
ax1.set_xlabel('Date')
ax1.set_ylabel('Close Price', color='tab:blue')
ax1.plot(data_apple['Date'], data_apple['Close'], label='Close Price', color='tab:blue')
ax1.tick_params(axis='y', labelcolor='tab:blue')

# Hacim grafiği
ax2 = ax1.twinx()
ax2.set_ylabel('Volume', color='tab:gray')
ax2.bar(data_apple['Date'], data_apple['Volume'], label='Volume', color='tab:gray', alpha=0.3)
ax2.tick_params(axis='y', labelcolor='tab:gray')

fig.tight_layout()
plt.title('AAPL Stock Price and Volume')
plt.show()
```

```python
# Hacim ve Kapanış Fiyatını Birlikte Görselleştirme

fig, ax1 = plt.subplots(figsize=(14, 7))

# Fiyat grafiği
sns.lineplot(x='Date', y='Close', data=data_apple, ax=ax1, color='b', label='Close Price')
ax1.set_ylabel('Close Price', color='b')
ax1.tick_params(axis='y', labelcolor='b')

# Hacim grafiği
ax2 = ax1.twinx()
sns.barplot(x='Date', y='Volume', data=data_apple, ax=ax2, alpha=0.3, color='gray')
ax2.set_ylabel('Volume', color='gray')
ax2.tick_params(axis='y', labelcolor='gray')

fig.tight_layout()
plt.title('AAPL Stock Price and Volume')
plt.show()
```

```python
# Kapanış Fiyatı Dağılımı.

plt.figure(figsize=(10, 6))
sns.histplot(data_apple['Close'], kde=True, bins=30)
plt.title('Distribution of AAPL Close Prices')
plt.xlabel('Close Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Kapanış Fiyatları ve Hacim Arasındaki İlişki

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Volume', y='Close', data=data_apple)
plt.title('AAPL Close Price vs. Volume')
plt.xlabel('Volume')
plt.ylabel('Close Price')
plt.show()
```

```python
# Hareketli Ortalama ile Zaman Serisi.

data_apple['SMA_20'] = data_apple['Close'].rolling(window=20).mean()

plt.figure(figsize=(14, 7))
sns.lineplot(x='Date', y='Close', data=data_apple, label='Close Price')
sns.lineplot(x='Date', y='SMA_20', data=data_apple, label='20-Day SMA', color='orange')
plt.title('AAPL Stock Price with 20-Day Simple Moving Average')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
# işlem hacmini (Volume) farklı zaman dilimlerine göre bir pasta grafikte


# Her ayın toplam işlem hacmini hesaplama
data_apple['Month'] = data_apple['Date'].dt.strftime('%B')
monthly_volume = data_apple.groupby('Month')['Volume'].sum()

# Pasta grafiği çizme
plt.figure(figsize=(10, 6))
plt.pie(monthly_volume, labels=monthly_volume.index, autopct='%1.1f%%', startangle=140)
plt.title('Monthly Trading Volume Distribution')
plt.axis('equal')
plt.show()
```

```python
# Her ayın ortalama kapanış fiyatlarını bir çubuk grafikte göstermek.

# Her ayın ortalama kapanış fiyatını hesaplama
monthly_close = data_apple.groupby('Month')['Close'].mean()

# Çubuk grafiği çizme
plt.figure(figsize=(10, 6))
sns.barplot(x=monthly_close.index, y=monthly_close.values, palette='viridis')
plt.title('Average Monthly Close Price')
plt.xlabel('Month')
plt.ylabel('Average Close Price')
plt.xticks(rotation=45)
plt.show()
```

```python
# Her ayın kapanış fiyatlarının dağılımını bir kutu grafikte göstermek.

plt.figure(figsize=(14, 7))
sns.boxplot(x='Month', y='Close', data=data_apple, palette='viridis')
plt.title('Monthly Close Price Distribution')
plt.xlabel('Month')
plt.ylabel('Close Price')
plt.xticks(rotation=45)
plt.show()
```

```python
# Fiyatların bir korelasyon matrisi ile ısı haritasında göstermek.

# Kapanış fiyatı ve işlem hacmi arasındaki korelasyonu hesaplama
correlation_matrix = data_apple[['Close', 'Volume']].corr()

# Isı haritası çizme
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Hareketli ortalamaları içeren bir çizgi grafiği.

data_apple['SMA_20'] = data_apple['Close'].rolling(window=20).mean()
data_apple['SMA_50'] = data_apple['Close'].rolling(window=50).mean()

plt.figure(figsize=(14, 7))
sns.lineplot(x='Date', y='Close', data=data_apple, label='Close Price')
sns.lineplot(x='Date', y='SMA_20', data=data_apple, label='20-Day SMA', color='orange')
sns.lineplot(x='Date', y='SMA_50', data=data_apple, label='50-Day SMA', color='green')
plt.title('AAPL Stock Price with Moving Averages')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```
